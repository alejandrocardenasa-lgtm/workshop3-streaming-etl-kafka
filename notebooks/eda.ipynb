```python
datasets = {
    "2015": df_2015,
    "2016": df_2016,
    "2017": df_2017,
    "2018": df_2018,
    "2019": df_2019
}

for year, df in datasets.items():
    print(f"\n===== {year} COLUMNS =====")
    print(df.columns.tolist())
```

# Schema Analysis
The datasets from 2015 to 2019 do not share a fully consistent schema.

Main inconsistencies identified:

- Different column naming conventions
- Different punctuation styles
- Different confidence interval columns
- Missing region information in newer datasets
- Different GDP and health column names
- Rank column naming inconsistencies

A unified schema is required before integrating all datasets into a single analytical dataset.

# Proposed Unified Schema
Based on the schema comparison across all datasets, the following unified analytical schema is proposed:

- country
- year
- happiness_rank
- happiness_score
- gdp
- family
- health
- freedom
- generosity
- corruption

These columns were selected because they are consistently available across most datasets and are relevant for machine learning prediction tasks.

```python
# Outlier Detection

numeric_columns = [
    'Happiness Score',
    'Economy (GDP per Capita)',
    'Family',
    'Health (Life Expectancy)',
    'Freedom',
    'Generosity'
]

df_2015[numeric_columns].boxplot(figsize=(12,6))
plt.title("2015 Numerical Features - Outlier Detection")
plt.xticks(rotation=45)
plt.show()
```

# Data Quality Observations

1. No duplicated records were found in any dataset.

2. Significant schema inconsistencies were identified across years:
   - Different column names
   - Different naming conventions
   - Different punctuation styles
   - Different confidence interval columns

3. Most datasets contain complete numerical information.

4. The 2018 dataset contains one missing value in the
   "Perceptions of corruption" column.

5. Numerical data types are generally consistent across datasets.

6. Several columns represent the same concept but use different names.
   Example:
   - "Economy (GDP per Capita)"
   - "Economy..GDP.per.Capita."
   - "GDP per capita"

7. Some columns are unnecessary for the machine learning task and may
   be removed during transformation:
   - Standard Error
   - Confidence Intervals
   - Whisker columns
   - Dystopia Residual

8. Boxplot analysis suggests the presence of some outliers, particularly
   in generosity and family-related features, but they appear to be
   valid country-level variations rather than data quality errors.
